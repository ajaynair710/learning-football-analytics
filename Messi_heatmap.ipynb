```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
import seaborn as sns
```

```python
#Read in the data
df = pd.read_csv('messibetis.csv')
```

```python
#convert the data to match the mplsoccer statsbomb pitch
df['x'] = df['x']*1.2
df['y'] = df['y']*.8
df['endX'] = df['endX']*1.2
df['endY'] = df['endY']*.8
```

```python
df
```

```python
# Create the figure and axis
fig, ax = plt.subplots(figsize=(13.5, 8))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

# Create the pitch
pitch = Pitch(pitch_type='statsbomb',
              pitch_color='#22312b', line_color='#c7d5cc')  # Removed tight_layout

# Draw the pitch on the ax figure and invert the axis for this specific pitch
pitch.draw(ax=ax)
plt.gca().invert_yaxis()

#Create the heatmap
kde = sns.kdeplot(
    data=df,
    x='x',
    y='y',
    shade=True,
    alpha=.5,
    n_levels=10,
    cmap='magma'
)

# Add the title
plt.title('Messi Heat Map vs Real Betis', color='white', size=20)

# Optimize layout
fig.tight_layout()

plt.xlim(0,120)
plt.ylim(0,80)

# Show the plot
plt.show()
```
